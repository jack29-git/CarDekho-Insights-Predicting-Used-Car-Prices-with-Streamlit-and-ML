```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import pickle
```

```python
data = pd.read_csv('Total_cars_data.csv')
data
```

## data cleaning

```python
# count number of missing values
data.isnull().sum()
```

```python
#drop columns from dataframe
data.drop(columns=['it','Gear Box','Drive Type'],inplace=True,axis=1)
data
```

```python
#body type unique values
data['bt'].unique()
```

```python
#retrive nan values from body type
data[pd.isna(data['bt'])]
```

```python
#replace nan values with model name
data.loc[data['model']=='Toyota Vellfire','bt']= 'Minivans'
data.loc[data['model']=='Maruti Eeco','bt']= 'Minivans'
```

```python
data['bt'].unique()
```

```python
#retrieve seats nan values
data[pd.isna(data['Engine'])]
```

```python
#drop nan values
data.dropna(subset=["Insurance Validity","Kms Driven","Engine","Color"],inplace=True)
```

```python
#retrieve seats nan values
data[pd.isna(data["Seats_1"])]
```

```python
#replace nan values of seats 
data.loc[data["model"]=="Maruti Estilo","Seats_1"]= 5.0
data.loc[data["model"]=="Nissan Kicks","Seats_1"]= 5.0
data.loc[data["model"]=="Ford Endeavour","Seats_1"]= 7.0
data.loc[data["model"]=="Ambassador","Seats_1"]= 5.0
```

```python
#again check for the missing values
data.isnull().sum()
```

```python
#drop column 'engine' from the dataframe
data.drop(columns=['Engine'],inplace=True)
data.reset_index(drop=True,inplace=True)
```

```python
data
```

```python
data['Kms Driven']= data['Kms Driven'].str.replace('Kms','').str.replace(',','').str.strip().astype(int)
data['Mileage']= data['Mileage'].str.replace('kmpl','').str.replace('km/kg','').str.strip().astype(float)
data['Seats_1']=data['Seats_1'].astype(int)
data['price']=data['price'].str.extract(r'(\d+\.?\d*)').astype(float)
```

```python
data.info()
```

```python
data['Mileage']=data['Mileage'].fillna(data['Mileage'].mean())
```

```python
data.isnull().sum()
```

```python
#Rename column names
data=data.rename(columns={'ft':'Fuel type','bt':'body type','Seats_1':'Seats','oem':'Brand'})
data
```

```python
#removing outliers from the 'price' column using IQR
#calculate Q1, Q3 and IQR
Q1 = data["price"].quantile(0.25)
Q3 = data["price"].quantile(0.75)
IQR = Q3-Q1

#filter out outliers from dataframe
df_filtered = data[(data["price"] >= (Q1 - 1.5 * IQR)) & (data["price"] <= (Q3 + 1.5 * IQR))]
#count of lower outliers
low_outliers = len(data[data['price'] < (Q1 - 1.5 * IQR)])
#count of upper outliers 
high_outliers = len(data[data['price'] > (Q3 + 1.5 * IQR)])
#print outliers counts
print(f"Count of lower outliers: {low_outliers}")
print(f"Count of upper outliers: {high_outliers}")
```

```python
df_filtered.reset_index(inplace=True,drop=True)
df_filtered
```

```python
#distribution of price after outliers removal
plt.figure(figsize=(10,6))
sns.histplot(x=data['price'],kde=True, bins=30)
plt.title('Distribution of price after outliers removal')
plt.xlabel('price')
plt.ylabel('frequency')
plt.show()
```

## scaling and encoding

```python
# creating a list to hold name of categorical columns
catg_col= df_filtered.select_dtypes(include="object").columns.to_list()
catg_col
```

```python
#one hot encoding
from sklearn.preprocessing import OneHotEncoder
#create the encoder
encoder = OneHotEncoder(sparse_output=False,handle_unknown="ignore")
#fit the encoder to the df_filtered
encoder.fit(df_filtered[catg_col])
```

```python
encoder.categories_
```

```python
#get the column names after encoding
new_encoded_col = list(encoder.get_feature_names_out(catg_col))
new_encoded_col
```

```python
#transforming the data
df_filtered[new_encoded_col]= encoder.transform(df_filtered[catg_col])
df_filtered.head()
```

```python
#drop the unwanted columns
df_filtered.drop(columns=catg_col,inplace=True)
df_filtered
```

```python
with open ('encoder.pkl','wb') as file:
    pickle.dump(encoder,file)
```
